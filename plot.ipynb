```python
def average(array,num = 20):
    array = array.reshape((-1,num))
    array = np.mean(array, axis = 1)

    return array.squeeze()
```

```python
fig, ax = plt.subplots(nrows =1, ncols = 2, figsize=  (12,6))

ax[0].plot(average(Ade5Lloss), "r--")
ax[0].plot(average(Ade6Lloss), "g--")
ax[0].plot(average(Ade7Lloss), "b--")
ax[0].plot(average(Lione5Lloss), "r")
ax[0].plot(average(Lione6Lloss), "g")
ax[0].plot(average(Lione7Lloss), "b")
ax[0].legend(['AdamW, lr = 3e-5','3e-6','3e-7','Lion, lr =3e-5','3e-6','3e-7'])
ax[0].set_title('Linear Schedule')

ax[1].plot(average(Ade5noloss), "r--")
ax[1].plot(average(Ade6noloss), "g--")
ax[1].plot(average(Ade7noloss), "b--")
ax[1].plot(average(Lione5noloss), "r")
ax[1].plot(average(Lione6noloss), "g")
ax[1].plot(average(Lione7noloss), "b")
ax[1].legend(['AdamW, lr = 3e-5','3e-6','3e-7','Lion, lr =3e-5','3e-6','3e-7'])
ax[1].set_title('No Schedule')
```
